## importing dataset with GDP growth

```python
# import GDP growth dataset
df_GDPgrowth = pd.read_csv('/content/drive/MyDrive/Birth rate and GDP growth analysis/gdp_growth.csv', header=0)
df_GDPgrowth.drop(columns='Unnamed: 65',inplace=True)
```

```python
# see the 20 highest countries with high growth rate
df_GDPgrowth.sort_values('2020',ascending=False, inplace=True)
top_20_GDPgrowth = df_GDPgrowth.head(20)
```

```python
# get the 20 lowest countries with GDP growth 
df_GDPgrowth_withoutnan = df_GDPgrowth[df_GDPgrowth['2020'].notna()].copy()
df_GDPgrowth_withoutnan.sort_values('2020', ascending=False, inplace=True)
bottom_20_GDPgrowth = df_GDPgrowth_withoutnan.tail(20)
```

```python
bottom_20_GDPgrowth
```

# Get the South Korea's fertility rate and GDP growth rate

```python
import matplotlib.pyplot as plt
```

```python
# Generate South Korea's fertility rate
df_fertility_Korea = df_fertility.loc[df_fertility['Country'] == 'South Korea']
df_fertility_Korea.set_index('Country',inplace=True)
```

```python
# Generate South Korea's GDP growth
df_GDPgrowth_Korea = df_GDPgrowth.loc[df_GDPgrowth['Country Name'] == 'Korea, Rep.']
df_GDPgrowth_Korea.set_index('Country Name',inplace=True)
```

```python
# drop the unnecessary column
df_GDPgrowth_Korea.drop(columns='Code',inplace=True)
```

```python
df_fertility_Korea
```

```python
df_GDPgrowth_Korea
```

```python
# concatenate two dataset
df_Korea = pd.concat([df_fertility_Korea, df_GDPgrowth_Korea], axis=0)
# replace nan value with recent 3 years of GDP growth
recent_3_years = df_Korea[['1961','1962','1963']]
mean_value = recent_3_years.mean(axis=1)
df_Korea['1960'].fillna(mean_value, inplace=True)
```

```python
# transpose the dataset
df_Korea = df_Korea.T
df_Korea
```

```python
# rename the column
df_Korea = df_Korea.rename(columns={'South Korea': 'South Korea fertility rate', 'Korea, Rep.': 'South Korea GDP growth rate'})
```

```python
df_Korea
```

```python
# adjust the size
fig = plt.figure(figsize=(10,7))

# Plot the fertility rate only
plt.scatter(df_Korea.index, df_Korea['South Korea GDP growth rate'])

# add labels and title
plt.xlabel("Year")
plt.ylabel("South Korea GDP growth rate")
plt.title("GDP growth rate in South Korea")

# show the plot
plt.show()
```

```python
# Let's see the two year with lowest GDP growth rate
two_lowest_values = df_Korea["South Korea GDP growth rate"].nsmallest(2)
print('Lowest year with GDP growth rate: ',two_lowest_values)
```

```python
# adjust the size
fig = plt.figure(figsize=(10,7))

# Plot the fertility rate only
plt.scatter(df_Korea.index, df_Korea['South Korea fertility rate'])

# add labels and title
plt.xlabel("Year")
plt.ylabel("South Korea fertility rate")
plt.title("South Korea fertility rate")

# show the plot
plt.show()
```

```python
# adjust the size
fig = plt.figure(figsize=(10,7))

# Plot the dataset
plt.scatter(df_Korea['South Korea fertility rate'], df_Korea['South Korea GDP growth rate'])

# add labels and title
plt.xlabel("South Korea fertility rate")
plt.ylabel("South Korea GDP growth rate")
plt.title("Fertility rate vs GDP growth rate in South Korea")

# show the plot
plt.show()
```

# China

```python
df_fertility_China = df_fertility.loc[df_fertility['Country']=='China']
df_fertility_China.set_index('Country',inplace=True)
```

```python
df_GDPgrowth_China = df_GDPgrowth.loc[df_GDPgrowth['Country Name']=='China']
df_GDPgrowth_China.set_index('Country Name',inplace=True)
df_GDPgrowth_China.drop(columns='Code',inplace=True)
```

```python
# Concatenate two data
df_China = pd.concat([df_fertility_China, df_GDPgrowth_China], axis=0)
# replace nan value with recent 3 years of GDP growth
recent_3_years = df_China[['1961','1962','1963']]
mean_value = recent_3_years.mean(axis=1)
df_China['1960'].fillna(mean_value, inplace=True)

df_China
```

```python
# transpose the dataset
df_China = df_China.T
df_China
```

```python
# rename the column
df_China.columns = ['China fertility rate','China GDP growth rate']
```

```python
|# adjust the size
fig = plt.figure(figsize=(10,7))

# Plot the fertility rate only
plt.scatter(df_China.index, df_China['China GDP growth rate'])

# add labels and title
plt.xlabel("Year")
plt.ylabel("China GDP growth rate")
plt.title("GDP growth rate in China")

# show the plot
plt.show()
```

```python
# adjust the size
fig = plt.figure(figsize=(10,7))

# Plot the fertility rate only
plt.scatter(df_China.index, df_China['China fertility rate'])

# add labels and title
plt.xlabel("Year")
plt.ylabel("China fertility rate")
plt.title("China fertility rate")

# show the plot
plt.show()
```

```python
# Let'see the period from 1990
# adjust the size
fig = plt.figure(figsize=(10,7))
# reset the index
df_China_1990 = df_China.reset_index(drop=False)
# change the type
df_China_1990['index'] = df_China_1990['index'].astype(int)
# Only remains from 1990~2020
df_China_1990 = df_China_1990.loc[df_China_1990['index'] >= 1990]

plt.scatter(df_China_1990['index'], df_China_1990['China fertility rate'])
plt.xlabel('Year')
plt.ylabel('Fertility Rate in China')
plt.title('Fertility Rate from 1990 to 2020')
plt.show()
```

```python
# adjust the size
fig = plt.figure(figsize=(10,7))

# Plot the dataset
plt.scatter(df_China['China fertility rate'], df_China['China GDP growth rate'])

# add labels and title
plt.xlabel("China fertility rate")
plt.ylabel("China GDP growth rate")
plt.title("Fertility rate vs GDP growth rate in China")

# show the plot
plt.show()
```
